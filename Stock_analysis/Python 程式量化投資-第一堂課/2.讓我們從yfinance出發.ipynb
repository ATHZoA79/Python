### <span style="color:red "> <算法二> 幾何收益率(對數收益率)-Log returns
-----


```python
# 以 2020-09-23 <-> 2020-09-24為例，算法如下
np.log(417.185669/420.553955)
```

```python
# 以 2020-09-24 <-> 2020-09-25為例，算法如下
np.log(418.472534/417.485565)
```

### <span style="color:red "> 兩種算法，有差別嗎??我們要用哪一種?

### <span style="color:red "> 算一下跨期的對數收益率

```python
# 以 2020-09-21 <-> 2020-09-25為例，算法如下
np.log(424/440)
```

```python
# 從20200921-20200925每天收益率加起來
np.log(424/423)+np.log(423/433.5)+np.log(433.5/437)+np.log(437/440)
```

### <span style="color:red "> 算一下跨期的簡單收益率

```python
# 以 2020-09-21 <-> 2020-09-25為例，算法如下
(424-440)/440
```

```python
(424-423)/423+(423-433.5)/433.5+(433.5-437)/437+(437-440)/440
```

#   讓我幫大家，把整段碼整理一下吧!

```python
import pandas as pd
import numpy as np
import yfinance as yf

#紅色是可以改的
#記得台股要加.tw
Target='2330.tw'

df= yf.download(
Target,
start='2022-2-7',
end='2023-2-7',
progress=False
)

########################################################################################################
df = df.reset_index(level=0)
df['Date'] = df['Date'].dt.tz_localize(None)
df.rename(columns={'Date':'日期','Open':'開盤價','High':'盤中最高價','Low':'盤中最低價','Close':'收盤價','Adj Close':'調整後收盤價','Volume':'成交量'}, inplace=True)
df['算術平均收益率'] = df.調整後收盤價.pct_change()
df['幾何收益率'] = np.log(df.調整後收盤價/df.調整後收盤價.shift(1))
df.to_excel(Target+'收益率計算.xlsx')
df
```
